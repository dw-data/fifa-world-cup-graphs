```python
import joypy
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
%matplotlib inline
```

```python
df=pd.read_csv(r'/Users/juliamerk/Documents/FIFA Story_Sports department/ageperposition.csv',delimiter=",")
```

```python
df.head()
```

```python
df1=pd.concat([df.iloc[:,:-1].idxmax(axis=1),df.iloc[:,-1]], axis=1)
```

```python
df1 = df1.rename(columns={0: 'Position', 'age': 'Age'})
```

```python
df1.head()
```

```python
fig, axes = joypy.joyplot(df1, by="Position", column="Age",figsize=(10,14), x_range = (15,45))
```
